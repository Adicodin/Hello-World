```python
# files to be used by user input
ans_key = input('Enter the official answer key file name =')
responses = input('Enter the name of file with your responses =')
attempt = input('Enter the attempt month of jee mains =')
```

```python
# store actual ans key given by nta

ans_phy=[]
ans_chem=[]
ans_math=[]
ans_phy_b=[]
ans_chem_b=[]
ans_math_b=[]
ans_dict = {'Physics':ans_phy, 'Chemistry':ans_chem, 'Mathematics':ans_math}
ans_des_dict = {'Physics':ans_phy_b, 'Chemistry':ans_chem_b, 'Mathematics':ans_math_b}
with open('{}.txt'.format(ans_key), 'r') as f:
    heading = f.readline()
    for k in range(3):
        if 'subject' in heading:
            ind = 2
            start_of_ans = (f.tell())
            subject_head = (f.readline()).strip().split('\t')[0]
            ans_lst = ans_dict[subject_head]
            ans_lst_b = ans_des_dict[subject_head]
            f.seek(start_of_ans)
        else:
            ind = 1
            if k == 0:
                ans_lst = ans_phy
                ans_lst_b = ans_phy_b
            elif k == 1:
                ans_lst = ans_chem
                ans_lst_b = ans_chem_b
            else:
                ans_lst = ans_math
                ans_lst_b = ans_math_b

        for i in range(0, 40):
            if i%2 != 0:
                f.readline()
            else:
                ans_lst.append((f.readline()).strip().split('\t')[ind:])
        for i in range(10):
            ans_lst_b.append((f.readline()).strip().split('\t')[ind:])


ans_obj = [ans_phy, ans_chem, ans_math]
ans_des = [ans_phy_b, ans_chem_b, ans_math_b]
```

```python
# function to sort nested list
def sublist(lst, ind):
    flag = 0
    while flag < len(lst)-1:
        flag = 0
        for i in range(0, len(lst)-1):
            if int(lst[i][ind])>int(lst[i+1][ind]):
                lst[i], lst[i+1] = lst[i+1], lst[i]
            else:
                flag += 1
    return lst
```

```python
# store my responses, options data and store data in specific format ques, ques id, etc in csv_data, desc data csv upload
with open('{}.txt'.format(responses), 'r') as f:
    phy = []
    f.readline()
    for i in range(20):
        ques = []
        for i in range(13):
            ques.append((f.readline()).strip())
        phy.append(ques)
    phy_b = []
    f.readline()
    for i in range(10):
        ques = []
        for i in range(5):
            ques.append((f.readline()).strip())
        phy_b.append(ques)
    chem = []
    f.readline()
    for i in range(20):
        ques = []
        for i in range(13):
            ques.append((f.readline()).strip())
        chem.append(ques)
    chem_b = []
    f.readline()
    for i in range(10):
        ques = []
        for i in range(5):
            ques.append((f.readline()).strip())
        chem_b.append(ques)
    math = []
    f.readline()
    for i in range(20):
        ques = []
        for i in range(13):
            ques.append((f.readline()).strip())
        math.append(ques)
    math_b = []
    f.readline()
    for i in range(10):
        ques = []
        for i in range(5):
            ques.append((f.readline()).strip())
        math_b.append(ques)
obj_data = [phy, chem, math]
des_data = [phy_b, chem_b, math_b]
# options storing -------------
options = []
for k in obj_data:
    for i in k:
        opt = [i[6][14:], i[7][14:], i[8][14:], i[9][14:], i[10][14:]]
        options.append(opt)
# options storing --------------


# csv data i.e., data stored as per specific format i.e., as per ques, ques id, chosen option, etc.
csv_data = []
for k in obj_data:
    for i in k:
        if i[-1][16:].isnumeric():
            optionid = (i[6+int(i[-1][16:])][14:])
            a = [i[0][2:], i[6][14:], i[11][9:], i[12][16:], optionid]
            csv_data.append(a)
        else:
            optionid = (i[-1][16:])
            a = [i[0][2:], i[6][14:], i[11][9:], i[12][16:], optionid]
            csv_data.append(a)

csv_data_d = []
# heading_d = ['QUESTION', 'QUESTION ID', 'STATUS', 'GIVEN ANSWER']
for k in des_data:
    for i in k:
        a = [i[0][2:], i[3][14:], i[4][9:], i[1][15:]]
        csv_data_d.append(a)

```

```python
# editiong the specific format data by adding new columns, correct option, right/wrong, etc. and calculating marks

# this is for OBJ only
lst = csv_data
sublist(lst, 1) # to arrange as per question id
sublist(options, 0) # to arrange as per question id

phy_marks=chem_marks=math_marks=0
obj_marks = {'Phy Sec A':phy_marks, 'Chem Sec A':chem_marks, 'Maths Sec A':math_marks}

for i in range(0, len(lst)):
    if i<20:
        ans_lst = ans_phy
        ind = i
        subj = 'Phy Sec A'
    elif i in range(20, 40):
        ans_lst = ans_chem
        ind = i-20
        subj = 'Chem Sec A'
    else:
        ans_lst = ans_math
        ind = i-40
        subj = 'Maths Sec A'
    if lst[i][-1] == ans_lst[ind][1]:
        correct_option = options[i].index(ans_lst[ind][1])
        a0 = correct_option
        a1 = ans_lst[ind][1]
        a2 = 'Right'
        marks_given = 4
    else:
        correct_option = options[i].index(ans_lst[ind][1])
        a0 = correct_option
        a1 = ans_lst[ind][1]
        if lst[i][3].isnumeric():
            a2 = 'Wrong'
            marks_given = -1
        else:
            a2 = 'Unattempted'
            marks_given = 0
    obj_marks[subj] += marks_given
    lst[i].append(a0)
    lst[i].append(a1)
    lst[i].append(a2)
    lst[i].append(marks_given)
    lst[i].append(subj)

new_phy_data = sublist(lst[:20], 0)
new_chem_data = sublist(lst[20:40], 0)
new_math_data = sublist(lst[40:], 0)
new_obj_data = [new_phy_data, new_chem_data, new_math_data]


# this is only for DES data
lstd = csv_data_d
sublist(lstd, 1)

phy_b_marks=chem_b_marks=math_b_marks=0
des_marks = {'Phy Sec B':phy_b_marks, 'Chem Sec B':chem_b_marks, 'Maths Sec B':math_b_marks}

for i in range(0, len(lstd)):
    if i<10:
        ans_lst = ans_phy_b
        ind = i
        subj = 'Phy Sec B'
    elif i in range(10, 20):
        ans_lst = ans_chem_b
        ind = i-10
        subj = 'Chem Sec B'
    else:
        ans_lst = ans_math_b
        ind = i-20
        subj = 'Maths Sec B'
    if lstd[i][-1] == ans_lst[ind][1]:
            correct_answer = ans_lst[ind][1]
            a1 = 'Right'
            marks_given = 4
    else:
        correct_answer = ans_lst[ind][1]
        marks_given = 0
        if lstd[i][3].isnumeric():
            a1 = 'Wrong'
        else:
            a1 = 'Unattempted'
    des_marks[subj] += marks_given
    lstd[i].append(correct_answer)
    lstd[i].append(a1)
    lstd[i].append(marks_given)
    lstd[i].append(subj)

new_phy_d_data = sublist(lstd[:10], 0)
new_chem_d_data = sublist(lstd[10:20], 0)
new_math_d_data = sublist(lstd[20:], 0)
new_des_data = [new_phy_d_data, new_chem_d_data, new_math_d_data]
```

```python
# final uploading of obj data into csv
import csv
with open("obj_data_{}.csv".format(attempt), "w", newline='') as f:
    wobj = csv.writer(f)
    header = ['QUESTION', 'QUESTION ID', 'STATUS', 'CHOSEN OPTION', 'CHOSEN OPTION ID', 'CORRECT ANS OPTION', 'CORRECT ANS OPTION ID', 'ANSWER STATUS', 'MARKS ALLOTED', 'SUBJECT']
    wobj.writerow(header)
    for i in new_obj_data:
        wobj.writerows(i)
        wobj.writerow('\n')

with open("des_data_{}.csv".format(attempt), 'w', newline='') as f:
    wobj = csv.writer(f)
    heading_d = ['QUESTION', 'QUESTION ID', 'STATUS', 'GIVEN ANSWER', 'CORRECT ANSWER', 'ANSWER STATUS', 'MARKS ALLOTED', 'SUBJECT']
    wobj.writerow(heading_d)
    for i in new_des_data:
        wobj.writerows(i)
        wobj.writerow('\n')
```

```python
# final marks calculation and display

total_marks = 0
for i in obj_marks, des_marks:
    for j in i:
        total_marks += i[j]


print('\t\tWelcome!!! Your marks are below\n')
print('--------------------Physics--------------------\n')
print('Physics Sec A = {} marks\t\t\tPhysics Sec B = {} marks\n'.format(obj_marks['Phy Sec A'],des_marks['Phy Sec B']))
print('Physics total =', obj_marks['Phy Sec A']+des_marks['Phy Sec B'])
print('\n\n--------------------Chemistry--------------\n')
print('Chemistry Sec A = {} marks\t\t\tChemistry Sec B = {} marks\n'.format(obj_marks['Chem Sec A'],des_marks['Chem Sec B']))
print('Chemistry total =', obj_marks['Chem Sec A']+des_marks['Chem Sec B'])
print('\n\n--------------------Maths-------------------\n')
print('Maths Sec A = {} marks\t\t\tMaths Sec B = {} marks\n'.format(obj_marks['Maths Sec A'],des_marks['Maths Sec B']))
print('Maths total =', obj_marks['Maths Sec A']+des_marks['Maths Sec B'])
print()
print('TOTAL MARKS out of 300 =', total_marks)
```
